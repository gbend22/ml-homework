```python
import pandas as pd

df_test = pd.read_csv('C:/Users/Acer/Desktop/ml-homework/Data/test.csv')
```

```python
import mlflow
import pandas as pd
mlflow.set_tracking_uri("https://dagshub.com/gbend22/ml-homework.mlflow")
logged_model = 'models:/LinearRegression/3'

loaded_model = mlflow.sklearn.load_model(logged_model)
X_kaggle_test = df_test.drop(['Id'], axis=1)
X_kaggle_test
```

```python
predictions = loaded_model.predict(X_kaggle_test)

submission = pd.DataFrame({
    'Id': df_test['Id'],
    'SalePrice': predictions
})
submission.to_csv('submission.csv', index=False)
submission
```
